```python
proc_cost = pd.DataFrame()
for i in range(17):
    clean = pd.read_csv('cleaned_{}.csv'.format(str(i)))
    g = clean.groupby(['TMA_Acct','claim_type','proc_code','proc_code_format'])['eacl_prv_alcrg_at'].sum().reset_index()
    proc_cost = pd.concat([proc_cost,g],axis = 0)
```

```python
hcpc_doc = pd.read_excel('HCPC2021_JAN_TRANS_ANWEB.xlsx',sheet_name = 0)
```

```python
def top_proc_cost(claim_type,cluster):
    cluster_selected = proc_cost.loc[proc_cost['TMA_Acct'].isin(tma_cluster_dict[cluster]),:]
    claim_type_selected = cluster_selected.loc[cluster_selected['claim_type'] == claim_type,:]
    top_cost_proc = claim_type_selected.groupby(['proc_code','proc_code_format'])['eacl_prv_alcrg_at'].sum().reset_index()
    top_cost_proc.sort_values(by ='eacl_prv_alcrg_at',ascending = False,inplace = True)
    merged_top_cost_proc = top_cost_proc.merge(hcpc_doc,how = 'left',left_on = 'proc_code',right_on = 'HCPC')
    return merged_top_cost_proc
```
