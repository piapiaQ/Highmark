```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
col = []
for i in range(1, 37):
    j = "_traj_Group_T" + str(i)
    col.append(j)
x = []
for i in range(36):
    x.append(i)
```

```python
data = pd.read_csv("final_3a_prob.csv")
data1 = data[data["_traj_Group"] == 1]
result1 = []
for i in col:
    correct = data1[data1[i] == 1]
    result1.append(len(correct)/len(data1))

data2 = data[data["_traj_Group"] == 2]
result2 = []
for i in col:
    correct = data2[data2[i] == 2]
    result2.append(len(correct)/len(data2))

data3 = data[data["_traj_Group"] == 3]
result3 = []
for i in col:
    correct = data3[data3[i] == 3]
    result3.append(len(correct)/len(data3))

data4 = data[data["_traj_Group"] == 4]
result4 = []
for i in col:
    correct = data4[data4[i] == 4]
    result4.append(len(correct)/len(data4))
fig = plt.figure(figsize=(8,5))
plt.plot(x,result1)
plt.plot(x,result2)
plt.plot(x,result3)
plt.plot(x,result4)
plt.legend(loc='upper center', labels=['group1', 'group2', 'group3', 'group4'])
plt.xlabel('Time')
plt.ylabel('Percentage of Correct Patients')
plt.title('Patients Converge Time')
plt.show()
```
